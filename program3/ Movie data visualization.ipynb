+ id：标识号
+ imdb_id：IMDB 标识号
+ popularity：在 Movie Database 上的相对页面查看次数
+ budget：预算（美元）
+ revenue：收入（美元）
+ original_title：电影名称
+ cast：演员列表，按 | 分隔，最多 5 名演员
+ homepage：电影首页的 URL
+ director：导演列表，按 | 分隔，最多 5 名导演
+ tagline：电影的标语
+ keywords：与电影相关的关键字，按 | 分隔，最多 5 个关键字
+ overview：剧情摘要
+ runtime：电影时长
+ genres：风格列表，按 | 分隔，最多 5 种风格
+ production_companies：制作公司列表，按 | 分隔，最多 5 家公司
+ release_date：首次上映日期
+ vote_count：评分次数
+ vote_average：平均评分
+ release_year：发行年份
+ budget_adj：根据通货膨胀调整的预算（2010 年，美元）

```python
import numpy as np
import pandas as pd
filename = './movies.csv'
movies_df = pd.read_csv(filename)
movies_df.describe()
```

```python
#将csv文件导出为excel文件
def csv_to_xlsx_pd():
    csv = pd.read_csv('new_movies.csv', encoding='utf-8')
    csv.to_excel('new_movies.xlsx', sheet_name='data')


if __name__ == '__main__':
    csv_to_xlsx_pd()
```

```python
# 由于一个电影由多个电影公司所制作，所以对电影公司这一列进行拆分。
split_genres = movies_df['genres'].str.split('|', expand=True)
split_genres['id'] = movies_df['id']
merged_back = movies_df.merge(split_genres)
melted = pd.melt(
    merged_back, id_vars=['id', 'release_year'], 
    value_vars=[0, 1, 2, 3, 4], value_name='genre'
).drop('variable', axis = 1
).dropna()

melted.to_csv('id_year_genre.csv', index=False)
```

```python
#将csv文件导出为excel文件
def csv_to_xlsx_pd():
    csv = pd.read_csv('id_year_genre.csv', encoding='utf-8')
    csv.to_excel('id_year_genre.xlsx', sheet_name='data')


if __name__ == '__main__':
    csv_to_xlsx_pd()
```

```python
# 将电影公司进行拆分，仅留下环球影业和派拉蒙影业
movies = pd.read_csv('movies.csv')
s = movies['production_companies'].str.split('|').apply(pd.Series, 1).stack()
s.index = s.index.droplevel(-1)
s.name = 'production'
movies = movies.join(s)
movies = movies[movies['production'].isin(["Universal Pictures", "Paramount Pictures"])]
movies.to_csv('only_companies.csv')
```

```python
#将csv文件导出为excel文件
def csv_to_xlsx_pd():
    csv = pd.read_csv('only_companies.csv', encoding='utf-8')
    csv.to_excel('only_companies.xlsx', sheet_name='data')


if __name__ == '__main__':
    csv_to_xlsx_pd()
```
